```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import animation
```

```python
D = 2 # 工具直径
rotation_speed = 6000 # 回転速度
feed_speed = 0.25 # 送り速度
theta_0 = 0 # 初期角度
R = D/2 # 工具半径

T = 10 # 時間幅（ｓ）
N = 100000 # Ｔをいくつに分けるか, default=100000
h = float(T/N)
```

```python
time = []
x = []
y = []

for i in range(1000):
  t = i * h
  theta = theta_0 + np.degrees((rotation_speed/60)*2*np.pi*t)
  print(theta)
  theta = np.radians(theta)
  x_tmp = R*np.cos(theta)+feed_speed*t
  y_tmp = R*np.sin(theta)

  print(x_tmp, y_tmp)
```
